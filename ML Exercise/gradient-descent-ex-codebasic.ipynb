```python
x = df.iloc[:, 1].values
y = df.iloc[:, 2].values
print(type(x))
print(y)
```

***Intercept(b) & coefficient(m) using gradient descent & cost function***

```python
def gradient_descent(x,y):
    m_curr = 0
    b_curr = 0
    iterations = 1000000
    learning_rate = 0.000211
    n = len(x)
    cost_prev = 0
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = sum([val**2 for val in y-y_predicted])/n
        if math.isclose(cost, cost_prev, rel_tol=1e-20):
            break
        cost_prev = cost
        dm = -(2/n) * sum(x * (y-y_predicted))
        db = -(2/n) * sum(y-y_predicted)
        m_curr = m_curr - learning_rate * dm
        b_curr = b_curr - learning_rate * db
        if i%1000 == 0:
            print('m {}        b {}          cost {}       iterations {}'.format(m_curr, b_curr, cost, i))
        

gradient_descent(x,y)
```

***m & b using sklearn library***

```python
def predict_using_sklearn(x, y):
    from sklearn.linear_model import LinearRegression
    x = np.reshape(x, (-1,1))
    regressor = LinearRegression()
    regressor.fit(x,y)
    print('Using sklearn  m {}   &  b {}'.format(regressor.coef_, regressor.intercept_))
predict_using_sklearn(x,y)
```
